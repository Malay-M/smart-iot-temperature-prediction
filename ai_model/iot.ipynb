```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.CSV')
df.head()
```

```python
# Make a copy to avoid issues
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# Extract datetime features
df['month'] = df['time'].dt.month
df['day'] = df['time'].dt.day
df['hour'] = df['time'].dt.hour
df['minute'] = df['time'].dt.minute
df['second'] = df['time'].dt.second
df['day_of_week'] = df['time'].dt.dayofweek  # Monday=0, Sunday=6

# Drop year to avoid overfitting
df['time_no_year'] = df['time'].dt.strftime("%m-%d %H:%M:%S")

df = df.drop(columns=['time'])  # optional if you want only processed features

df.head()
```

```python
df = df.drop(columns=['pressure', 'lux', 'time_no_year'])
df.head()
```

```python

df["humidity"] = df["humidity"] + 4
old_min, old_max = df["temperature"].min(), df["temperature"].max()
new_min, new_max = 10, 50

df["temperature"] = (df["temperature"] - old_min) * (new_max - new_min) / (old_max - old_min) + new_min
```

```python
df.describe()
```

```python
df["target_temp"] = df["temperature"].shift(-1)
df = df.dropna()
```

```python
X = df.drop(columns=["target_temp"])
y = df["target_temp"]
```
